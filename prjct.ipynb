```python
from sklearn.model_selection import train_test_split 
data_train,data_test=train_test_split(data_use,test_size=0.2,random_state=13)
```

```python
X_train=data_train.drop(["tip"],axis=1)
y_train=data_train["tip"]
X_test=data_test.drop(["tip"],axis=1)
y_test=data_test["tip"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg=LinearRegression()
```

```python
lin_reg.fit(X_train,y_train)
```

```python
y_pred=lin_reg.predict(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd_regressor = SGDRegressor(max_iter=1000, tol=1e-3, random_state=42)
sgd_regressor.fit(X_train_scaled, y_train.ravel())
```

```python
y_pred = sgd_regressor.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```
